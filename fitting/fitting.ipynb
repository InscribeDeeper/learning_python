```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

## 安装包
# !conda install openpyxl -y
# !conda install matplotlib -y

def show_plot(mapping_function, x, y):
    yvals = mapping_function(x)

    plt.plot(x, y, '*',label='original values')
    plt.plot(x, yvals, 'r',label='polyfit values')
    plt.xlabel('x axis')
    plt.ylabel('y axis')
    plt.legend(loc=4) # 指定legend在图中的位置，类似象限的位置
    plt.title('polyfitting')
    plt.show()
```

## 读取数据

```python
dt = pd.read_excel("5200mAh放电曲线.xlsx", header=1) # 读取 文件
dt = dt.iloc[:-1] # 去除最后一个 异常值
dt[["时间", "电池电压\n（V）", "容量\n（mAh）"]].sort_values(by="电池电压\n（V）")
```

```python
dt.columns 
```

## 提取数据

```python
x = dt["容量\n（mAh）"].values
y = dt["电池电压\n（V）"].values
```

## 实验

```python
z1 = np.polyfit(x, y, 1) # 用1次多项式拟合，可改变多项式阶数； z1 是方程系数
mapping_function = np.poly1d(z1) #得到多项式函数 (根据 z1的系数)，按照阶数从高到低排列
print(mapping_function)  #显示多项式

yvals = mapping_function(x) # 预测 并且做可视化


plt.plot(x, y, '*',label='original values')
plt.plot(x, yvals, 'r',label='polyfit values')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc=4) # 指定legend在图中的位置，类似象限的位置
plt.title('polyfitting')
plt.show()
```

```python
degree = 2
z1 = np.polyfit(x, y, degree) # 用1次多项式拟合，可改变多项式阶数； z1 是方程系数
mapping_function = np.poly1d(z1) #得到多项式函数 (根据 z1的系数)，按照阶数从高到低排列
print(mapping_function)  #显示多项式

yvals = mapping_function(x) # 预测 并且做可视化
show_plot(mapping_function, x, y)
```
